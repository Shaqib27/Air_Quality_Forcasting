```python
import pandas as pd
import os

folder_path = "AirData"

dfs = []

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        
        dfs.append(df)

# Merge all CSVs
air_df = pd.concat(dfs, ignore_index=True)

print(air_df.shape)
air_df.head()


drop_col = [
    'MP-Xylene (ug/m3)',
    'Eth-Benzene (ug/m3)',
    'Xylene (ug/m3)',
    'RF (mm)',
    'SR (W/mt2)'
]

df.drop(columns=drop_col, inplace=True)

dcol = [ 'CO (mg/m3)','RH (%)','AT (degree C)','NO (ug/m3)']
df.drop(columns=dcol,inplace=True)

num_cols = [
    'PM2.5 (ug/m3)', 'PM10 (ug/m3)', 'NO2 (ug/m3)', 'NOx (ppb)',
    'NH3 (ug/m3)', 'SO2 (ug/m3)', 'Ozone (ug/m3)',
    'Benzene (ug/m3)', 'Toluene (ug/m3)',
    'Temp ()', 'WS (m/s)', 'WD (degree)', 'BP (mmHg)'
]

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

    df = df.rename(columns={
    'From Date': 'from_date',
    'To Date': 'to_date',
    'PM2.5 (ug/m3)': 'pm25',
    'PM10 (ug/m3)': 'pm10',
    'NO2 (ug/m3)': 'no2',
    'NOx (ppb)': 'nox',
    'NH3 (ug/m3)': 'nh3',
    'SO2 (ug/m3)': 'so2',
    'Ozone (ug/m3)': 'ozone',
    'Benzene (ug/m3)': 'benzene',
    'Toluene (ug/m3)': 'toluene',
    'Temp ()': 'temperature',
    'WS (m/s)': 'wind_speed',
    'WD (degree)': 'wind_direction',
    'BP (mmHg)': 'pressure'
})
df['from_date']= pd.to_datetime(df['from_date'])

df = df.sort_values('from_date').reset_index(drop=True)



```
